```python
#Task4
import numpy as npfrom 
from scipy import misc 
import matplotlib.pyplot as plt
import matplotlib.image as img
from scipy import ndimage as nimg
from scipy import misc 
from skimage import feature
import numpy as np

pic2 = misc.imread('mv.jpg',mode='L')
pic = img.imread ('mv.jpg')

edges1 = feature.canny(pic2)
edges2 = feature.canny(pic2, sigma=3)

fig = plt.figure(figsize=(15,10))
plt.subplot(131)
plt.imshow(pic[:,:])
plt.subplot(132)
plt.imshow(edges1, cmap='Reds')
plt.subplot(133)
plt.imshow(edges2, cmap='Reds')
plt.show()
```
